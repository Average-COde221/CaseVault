Preprocessing Steps:

1. merged data from 3 jsonl files into 1 jsonl file
2. converted lists and dictionaries to strings in attributes to list of strings and dictionaries with lists of strings
3. removed punctuation and whitespaces from textual fields
4. one hot encoded the categorical columns

```python
import pandas as pd
import re
```

```python
dataset1=pd.read_json(r"D:\hackathons\CaseVault\recommendation-engine\Data\raw\train.jsonl",lines=True)
dataset2=pd.read_json(r"D:\hackathons\CaseVault\recommendation-engine\Data\raw\test.jsonl",lines=True)
dataset3=pd.read_json(r"D:\hackathons\CaseVault\recommendation-engine\Data\raw\train.jsonl",lines=True)
```

```python
df_total=pd.concat([dataset1,dataset2,dataset3],ignore_index=True) # merging datasets
```

```python
for col in df_total.columns: # for all columns in dataframe
    if df_total[col].apply(lambda x: isinstance(x, (list, dict))).any(): #if any of them are instances of list or dictionary
        print(f"Column '{col}' contains unhashable types.") #print that they are unhashable
```

```python
# Columns to convert
columns_to_convert = [
    'facts', 'applicants', 'defendants',
    'allegedly_violated_articles', 'violated_articles',
    'court_assessment_references', 'silver_rationales', 'gold_rationales'
]

# Convert lists/dictionaries to strings
df_total[columns_to_convert] = df_total[columns_to_convert].map(
    lambda x: str(x) if isinstance(x, (list, dict)) else x
)
```

```python
df_merged = df_total.drop_duplicates()
```

```python
df_merged = df_merged.drop("case_id",axis="columns")
```

```python
df_merged=df_merged.drop("case_no",axis="columns")
```

```python
df_merged.isna().sum()
```

```python
df_merged['judgment_date'] = pd.to_datetime(df_merged['judgment_date'], errors='coerce')
```

```python
# Text cleaning function
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()  # Convert to lowercase
    return text
```

```python
df_merged['title'] = df_merged['title'].apply(clean_text)
df_merged['facts'] = df_merged['facts'].apply(clean_text)
```

```python
df_merged
```

```python
df_copy1=df_merged.copy()
```

```python
columns_to_encode = ['applicants', 'defendants','allegedly_violated_articles','violated_articles',	'court_assessment_references',	'silver_rationales'	,'gold_rationales']

df_encoded = pd.get_dummies(df_copy1, columns=columns_to_encode, drop_first=True)

print(df_encoded.head())
```
